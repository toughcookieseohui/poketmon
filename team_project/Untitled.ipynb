```python
from poketmon.models import *
```

```python
monsters = monster.objects.all()
```

```python
types = ''
for monster in monsters:
    types += monster.type
    types +=';'
print(types)
```

```python
list_types = types.replace(';', ',').split(',')
print(list_types)
```

```python
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
counter = Counter(list_types)
```

```python
print(counter)
```

```python
del counter['']
```

```python
wordcloud = WordCloud(
    font_path = "C:\Windows\Fonts\H2MJSM.ttf",
    background_color="white",
    width=1000,
    height=1000,
)

img = wordcloud.generate_from_frequencies(counter)
plt.imshow(img)
plt.savefig('cloud_types_dist.png')
```

```python
from matplotlib import font_manager, rc

font_path = "C:\Windows\Fonts\H2MJSM.ttf" =
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

labels = counter.keys()
sizes = counter.values()

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax.axis('equal')  # 원이 타원이 되지 않도록 설정

# 차트 제목 추가 (원하는 제목으로 대체하세요)
plt.title("포켓몬 타입 분포")

# 차트 보여주기
# plt.show()

# 이미지 저장 (원하는 파일 이름 및 확장자로 대체)
plt.savefig('pie_types_dist.png')
```
